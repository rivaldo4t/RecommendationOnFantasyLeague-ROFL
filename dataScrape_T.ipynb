# Match Result field addition

```python
path = 'data/2016-17/players'
filenames = glob.glob(path + "/*/gw.csv")
for f in filenames:
    a = pd.read_csv(f)
    b = a['team_h_score'] - a['team_a_score']
    for i in a.index:
        if a.loc[i, 'was_home'] == True:
            if b[i] < 0:
                a.loc[i, 'match_result'] = -1
            elif b[i] == 0:
                a.loc[i, 'match_result'] = 0
            else:
                a.loc[i, 'match_result'] = 1
        else:
            if b[i] < 0:
                a.loc[i, 'match_result'] = 1
            elif b[i] == 0:
                a.loc[i, 'match_result'] = 0
            else:
                a.loc[i, 'match_result'] = -1
    a.to_csv(f[:-4] + "_match_result.csv")
```

# Training and Prediction

```python
features = ['assists', 'bonus', 'goals_scored', 'start_cost', 'end_cost','minutes', 'total_points']
```

```python
import glob
path = 'data/2016-17/players'
filenames = glob.glob(path + "/*/history.csv")
p_name = pd.read_csv('data/2016-17/player_idlist.csv')
fname = p_name['first_name']
sname = p_name['second_name']
pname = fname + '_' + sname
```

```python
import pandas as pd
player_avg_history = pd.DataFrame(index = pname, columns = features)
player_avg_history[:5]
```

```python
path = 'data/2016-17/players/'
for i in player_avg_history.index:
    f = path + i + '/history.csv'
    f2 = unicode(f, "utf-8") # a shot in the dark; bullseye
    a = pd.read_csv(f2)
    b = a.mean(axis=0)
#     print b[features]
    player_avg_history.loc[i] = b[features]
    player_avg_history.loc[i, 'points_per_min'] = 0 if b['minutes'] == 0 else b['total_points'] / b['minutes']
```

```python
player_avg_history[:10]
```

```python
trainFeature = features[:3]
trainFeature.append(features[5])
x = player_avg_history[trainFeature]
y = player_avg_history['total_points']
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(x, y)

path = 'data/2016-17/players/'
player_score = {}
for i in player_avg_history.index:
#     print i
    f = unicode(path + i + '/gw.csv', "utf-8")
    a = pd.read_csv(f)
    a = a[a['minutes']>0]
    b = a[trainFeature][-5:]
#     print b
    c = regr.predict(b).mean() if b.shape[0] > 0 else 0
#     print c
    player_score[i] = c

import operator
sorted_x = sorted(player_score.items(), key=operator.itemgetter(1), reverse=True)
sorted_x[:10]
```

```python
from sklearn.svm import SVR
import numpy as np
clf = SVR(C=1.0, epsilon=0.2)
clf.fit(x, y)

path = 'data/2016-17/players/'
player_score = {}
for i in player_avg_history.index:
#     print i
    f = unicode(path + i + '/gw.csv', "utf-8")
    a = pd.read_csv(f)
    a = a[a['minutes']>0]
    b = a[trainFeature][-5:]
#     print b
    c = clf.predict(b).mean() if b.shape[0] > 0 else 0
#     print c
    player_score[i] = c
    
sorted_x = sorted(player_score.items(), key=operator.itemgetter(1), reverse=True)
sorted_x[:10]
```

```python
clf2 = linear_model.BayesianRidge()
clf2.fit(x,y)

path = 'data/2016-17/players/'
player_score = {}
for i in player_avg_history.index:
#     print i
    f = unicode(path + i + '/gw.csv', "utf-8")
    a = pd.read_csv(f)
    a = a[a['minutes']>0]
    b = a[trainFeature][-5:]
#     print b
    c = clf2.predict(b).mean() if b.shape[0] > 0 else 0
#     print c
    player_score[i] = c
    
sorted_x = sorted(player_score.items(), key=operator.itemgetter(1), reverse=True)
sorted_x[:10]
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(x, y)

path = 'data/2016-17/players/'
player_score = {}
for i in player_avg_history.index:
#     print i
    f = unicode(path + i + '/gw.csv', "utf-8")
    a = pd.read_csv(f)
    a = a[a['minutes']>0]
    b = a[trainFeature][-5:]
#     print b
    c = regr.predict(b).mean() if b.shape[0] > 0 else 0
#     print c
    player_score[i] = c

sorted_x = sorted(player_score.items(), key=operator.itemgetter(1), reverse=True)
sorted_x[:10]
```

```python
from sklearn import neural_network
clf3 = neural_network.MLPRegressor(
    hidden_layer_sizes=(10,),  activation='relu', solver='adam', alpha=0.001, batch_size='auto',
    learning_rate='constant', learning_rate_init=0.01, power_t=0.5, max_iter=1000, shuffle=True,
    random_state=9, tol=0.0001, verbose=False, warm_start=False, momentum=0.9, nesterovs_momentum=True,
    early_stopping=False, validation_fraction=0.1, beta_1=0.9, beta_2=0.999, epsilon=1e-08)

clf3.fit(x,y)

path = 'data/2016-17/players/'
player_score = {}
for i in player_avg_history.index:
#     print i
    f = unicode(path + i + '/gw.csv', "utf-8")
    a = pd.read_csv(f)
    a = a[a['minutes']>0]
    b = a[trainFeature][-5:]
#     print b
    c = regr.predict(b).mean() if b.shape[0] > 0 else 0
#     print c
    player_score[i] = c

sorted_x = sorted(player_score.items(), key=operator.itemgetter(1), reverse=True)
sorted_x[:10]
```
