```python
path = 'data/2016-17/players'
filenames = glob.glob(path + "/*/history.csv")
all_p_hData = (pd.read_csv(f).mean() for f in filenames)
p_hData = pd.concat(all_p_hData, keys=[i for i in range(683)])
print p_hData
```

```python
mat = np.zeros((683,9))
for i in range(683):
    mat[i][0] = p_hData[i]['assists']
    mat[i][1] = p_hData[i]['goals_scored']
    mat[i][2] = p_hData[i]['bonus']
    mat[i][3] = p_hData[i]['clean_sheets']
    mat[i][4] = p_hData[i]['creativity']
    mat[i][5] = p_hData[i]['saves']
    mat[i][6] = p_hData[i]['influence']
    mat[i][7] = p_hData[i]['start_cost']
    mat[i][8] = 0 if p_hData[i]['minutes'] == 0 else p_hData[i]['total_points'] / p_hData[i]['minutes']
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression(normalize=True)
regr.fit(mat[:,0:7], mat[:,8])
```

```python
print regr.coef_
print regr.intercept_
```

```python
p_name = pd.read_csv('data/2016-17/player_idlist.csv')
```

```python
fname = p_name['first_name']
sname = p_name['second_name']
pname = fname + '_' + sname
```

```python
import pandas as pd
player_avg_history = pd.DataFrame(index = pname, columns = features)
player_avg_history[:5]
```

```python
path = 'data/2016-17/players/'
for i in player_avg_history.index:
    f = path + i + '/history.csv'
    f2 = unicode(f, "utf-8") # a shot in the dark; bullseye
    a = pd.read_csv(f2)
    b = a.mean(axis=0)
#     print b[features]
    player_avg_history.loc[i] = b[features]
    player_avg_history.loc[i, 'points_per_min'] = 0 if b['minutes'] == 0 else b['total_points'] / b['minutes']
```

```python
player_avg_history
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = player_avg_history[features[:5]]
y = player_avg_history['points_per_min']
regr.fit(x, y)
```
